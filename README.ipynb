### Project Overview

This project uses data mining and machine learning methods to examine customer decline in the telecom sector. When consumers stop using a business, it's known as customer churn, and it can have a big impact on sales and long-term business expansion.

This project aims to identify customers who are at risk of switching telecom service providers by analysing past customer data and creating predictive models. Organisations can enhance customer satisfaction and implement focused retention strategies by comprehending the factors that lead to churn.

The entire business intelligence lifecycle—which includes problem definition, exploratory data analysis, predictive modelling, and strategic business recommendations—is followed in this project.

### Research Objectives

The project aims to answer the following key questions:

- Which factors most strongly influence customer churn?

- Can machine learning models accurately predict which customers are likely to leave?

- How can predictive insights be translated into actionable business strategies for customer retention?

### Dataset

The dataset used in this project contains 7,043 customer records and 33 attributes describing customer demographics, service subscriptions, billing information, and churn behaviour.


**Key Dataset Features**

 The dataset includes information such as:

 **Customer Information**
 - CustomerID
 - Gender
 - Location, etc.

**Service Information**
- Phone Service
- TV Streaming
- Movie Streaming, etc.

**Contract & Billing Information**
- Contract Type
- Monthly Charges
- Payment Method
- Total Charges, etc.

**Business Intelligence Variables**
- Customer Lifetime Value (CLTV)
- Churn Score

**Target Variables**
- Churn Label
- Churn Value
